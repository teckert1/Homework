```python
# -*- coding: UTF-8 -*-
"""PyRamen Homework Starter."""

# @TODO: Import libraries
import csv
import pandas as pd

# @TODO: Set file paths for menu_data.csv and sales_data.csv
#load menu data
menudf = pd.read_csv('menu_data.csv')
#preview menu data menudf.head()
menudf.head()

```

```python
#load sales data
salesdf = pd.read_csv('sales_data.csv')
#preview sales data salesdf.head()
salesdf.head()
```

```python
# @TODO: Initialize list objects to hold our menu and sales data
menu = []
sales = []

# @TODO: Read in the menu data into the menu list
#rename columns for merging
salesdf = salesdf.rename({'Menu_Item':'item'}, axis='columns')
salesdf


```

```python
#merge the data sets together
dfjoin = pd.merge(menudf,salesdf, on='item', how='inner')
dfjoin
    
```

```python
#remove unecessary columns 
dfjoin = dfjoin.drop(columns=['category','description','Line_Item_ID','Date','Credit_Card_Number'])
dfjoin
```

```python
#create a pivot table by menu item
import numpy as np
dfpivot=pd.pivot_table(dfjoin,values=['price','Quantity','cost'],index=['item'],aggfunc=np.sum)
dfpivot
```

```python
#calulate profit of each menu item and create a new row in the table for profit
dfpivot['Profit'] = (dfpivot['price'] - dfpivot['cost'])
dfpivot
```

```python
#export to csv
dfpivot.to_csv('menu_report.csv')
```
